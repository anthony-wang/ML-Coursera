# Exercise 2 - Logistic Regression (aka Logit Model, Logistic Classification)

```julia
# Logit model, v0.4

# Part 2: compute cost and gradient
using CSV
data = CSV.read("/Users/kevinliu/Documents/machine-learning-ex2/ex2/ex2data1.txt", datarow=1);

# Add X to a column of ones to compute y-intercept
X = hcat(ones(100,1), Matrix(data[:, [1,2]])); y = Vector(data[:, 3]);

# Sigmoid function
function sigmoid(z)
    1.0 ./ (1.0 .+ exp.(-z))
end

sigmoid(0) # => 0.5
z = rand(3,1); sigmoid(z) # vector
z = rand(3,3); sigmoid(z) # matrix

# Hypothesis: linearly combines X[i] and θ[i], to calculate all instances of cost()
function h(θ, X)
    z = 0
    for i in 1:length(θ)
        z += θ[i] .* X[i, :]
    end
    sigmoid(z)
end

h([-24, 0.2, 0.2], X)

# Vectorized cost function
function cost(θ, X, y)
    hx = sigmoid(X * θ)
    m = length(y)
    global J = (-y' * log.(hx) - (1 - y') * log.(1 - hx)) / m
    global grad = X' * (hx - y) / m
    println("Cost is $J")
    println("Gradient is $grad")
end

cost([0, 0, 0], X, y)
# Cost is 0.693147180559945
# Gradient is [-0.1, -12.0092, -11.2628]
# as expected

cost([-24, 0.2, 0.2], X, y)
# Cost is 0.21833019382659777
# Gradient is [0.042903, 2.56623, 2.6468]
# as expected
```

### Optimizing J and gradient

```julia
# Part 3: optimize J and gradient
using Optim
# minimize J2
J2(θ) = (-y' * log.(sigmoid(X * θ)) - (1 - y') * log.(1 - sigmoid(X * θ))) / length(y)
optimize(J2, zeros(3), BFGS())

thetas = Optim.minimizer(optimize(J2, zeros(3), ConjugateGradient()))
# => 3-element Array{Float64,1}:
#     -25.1614
#       0.206232
#       0.201472
# thetas as expected
```

```julia
Optim.minimum(optimize(J2, zeros(3), BFGS()))
# => 0.2034977015894402
# θ for J as expected
```

### Accurately predicting

```julia
# Part 4: accurately predict
# admittance probability with 45% on exam 1 and 85% on exam 2
prob = sigmoid([1 45 85] * θ) # => 0.776287 as expected

# training set accuracy
function predict(θ, X)
    m = size(X, 1)
    p = zeros(m, 1)
    for i = 1:m
        if sigmoid(X * θ)[i] >= 0.5
           p[i] = 1
        else
           p[i] = 0
        end
    end
    p
end

p = predict(θ, X)

# dot convert to convert a vector
convert.(Int64, p)

# train accuracy
# dot the == operator for a vector of trues and falses
mean(p .== y) # => 0.89 as expected
```
